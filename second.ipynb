# Scatter plot

```python

x_data = df['Number_of_Drivers']
y_data = df['Average_Ratings']

plt.figure(figsize=(10, 6))
plt.scatter(x_data, y_data, alpha=0.6, color='darkred') 

# à¤²à¥‡à¤¬à¤² à¤”à¤° à¤Ÿà¤¾à¤‡à¤Ÿà¤² à¤¦à¥‡à¤‚
plt.xlabel(" Number_of_Drivers")
plt.ylabel("Average_Ratings")
plt.title("Scatter Plot: Relationship between two variables")
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

### Most locations maintain an average rating between 3.5 and 5.0, irrespective of whether they have a low (around 5) or high (around 80) number of drivers.


# Bar Chart


```python

plot_data = df.groupby('Customer_Loyalty_Status')['Number_of_Riders'].mean().sort_values(ascending=False).reset_index()


plt.figure(figsize=(5, 3))
plt.bar(x=plot_data['Customer_Loyalty_Status'], 
        height=plot_data['Number_of_Riders'], 
        color='skyblue')

plt.xlabel("Customer_Loyalty_Status")
plt.ylabel("Number_of_Riders")
plt.title("Bar Chart: Average Value by Category")
plt.xticks(rotation=0, ha='right') 
plt.tight_layout()
plt.show()
```
